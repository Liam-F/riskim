```python
from datagate import *
from datetime import datetime
from util import ConfigLoader
```

```python
config_loader = ConfigLoader('backtest.ini')
fund_file_path = config_loader.system.fund_path
g_start, g_end = config_loader.experiment.general_start_end
```

```python
datahandler = DataHandler(CSVLoader(fund_file_path), g_start, g_end)
symbol_list = datahandler.symbol_list
print "the num of valid symbols in the date range is %d." % len(symbol_list)
fund_dg = DataGate(datahandler, symbol_list)
```

```python
date_list = []
values_list = []
symbol_list = fund_dg.symbol_list
date_start = datetime(2015, 1, 1)

start_flag = False

for date, values in fund_dg:
    if date >= date_start and not start_flag:
        start_flag = True
        base_adjust = [1.0 / v for v in values]
    if start_flag:
        date_list.append(date)
        values_list.append([a * b for a, b in list(zip(values, base_adjust))])
```

```python
import pandas as pd
```

```python
pdf = pd.DataFrame(index=date_list, data=values_list, columns=symbol_list)
pdf.to_csv('sample_fund.csv')
```
